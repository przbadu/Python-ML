```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very imporatnt in Feature Selection

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## JointPlot

A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot

### Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

Based on above jointplot analysis, we can clearly tell couple of things:

- most of the peoples pay bill between 10-20
- most of the poeple give tips between 2-4
- There are very few people who give tips around 10, that also when the bill amount is around 50

```python
# reg: regression
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

## Pair plot

A `pair plot` is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
# we can also pairplot based on certain independet variable, using hue='' parameter
sns.pairplot(df, hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df, hue='smoker')
```

## Distplot

Distplot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. in this section we will discuss about

- boxplot
- violinplot
- countplot
- barplot

### Count plot

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot('time', data=df)
```

```python
sns.countplot('day', data=df)
```

### Bar plot

```python
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

### Box plot

A box an whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('sex', 'total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v')
```

## Violin Plot

Violin plot helps us to see both distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
